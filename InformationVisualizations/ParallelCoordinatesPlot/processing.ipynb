```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./output_consolidated.csv")
df.describe()
```

```python
df.head()
```

```python
country_df = df.groupby(['Country']).sum().reset_index()
country_df.set_index("Country", inplace=True)
country_df.drop(['F1', 'year'], axis=1, inplace=True)
country_df
```

```python
yearly_df = df.groupby('year').sum(numeric_only=True).reset_index()
yearly_df.set_index('year', inplace=True)
yearly_df.drop(['F1'], axis=1, inplace=True)
yearly_df
```

```python
yearly_df['colorVal'] = [0, 1000, 2000, 3000]
country_df['colorVal'] = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 11000, 12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000, 30000, 31000, 32000, 33000]
```

```python
yearly_df
```

```python
yearly_df.to_csv("year-wise.csv")
country_df.to_csv("country-wise.csv")
```
